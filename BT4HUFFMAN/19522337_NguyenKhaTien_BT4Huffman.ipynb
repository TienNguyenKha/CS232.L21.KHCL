##Huffman code:

Tạo node:

```python
class NodeTree(object):
    def __init__(self,left=None,right=None):
        self.left=left
        self.right=right
    def children(self):
        return (self.left,self.right)
```

Thêm mã 'encode' cho từng node:

```python
def huffman_encoding(node,binary=''):
    if type(node) is str:#nếu là node cao nhất or node lá
        return {node: binary}
    (l,r)=node.children()
    d=dict()
    d.update(huffman_encoding(l,binary+'0'))
    d.update(huffman_encoding(r,binary+'1'))
    return d
```

Tạo bảng tần suất xuất hiện của các phần tử và tạo cây:

```python
def table_of_char(text):    
    table={}
    #tính tần suất xuất hiện
    for i in text:
        if i in table:
            table[i]+=1
        else:
            table[i]=1#nếu dict ko có thì dict tự thêm phần tử đó vào và gán cho value là 1.
    TABLE=table#Bảng tần suất xuất hiện của các phần tử.
    table=sorted(table.items(),key=lambda x:x[1],reverse=True)
    nodes=table
    while len(nodes)>1:
        (k1,c1)=nodes[-1]
        (k2,c2)=nodes[-2]
        nodes=nodes[:-2]
        node=NodeTree(k1,k2)
        nodes.append((node,c1+c2))

        nodes=sorted(nodes,key=lambda x:x[1], reverse=True)
    huffman=huffman_encoding(nodes[0][0])#thêm mã encode cho từng node
    d=dict()
    for (char,frequency) in table:
        d.update({char:huffman[char]})
    return d,TABLE,huffman
```

Encode:

```python
def encoding(text,d):
    temp=''
    for i in text:
        temp+=d[i]
    return temp
```

Decode:

```python
def decoding(text,d):
    #reversed dict:
    d_reverse={}
    d_reverse_key=[]
    for i in tuple(d.items()):
        d_reverse.update({i[1]:i[0]})
        d_reverse_key.append(i[1])
    text=list(text)
    temp=text[0]
    decode=''
    for k in range(1,len(text)):
        if temp not in d_reverse_key:
            temp+=text[k]
        else:
            decode+=d_reverse[temp]
            temp=text[k]
    #decode thằng cuối
    decode+=d_reverse[temp]
    return decode
```

Tính toán hệ số nén:

```python
import math
```

```python
def hesonen(text,table,huffman):
  #table là 1 dict gồm Char và tần số 
  #huffman là 1 dict gồm Char và encode của nó
  items_frequence=list(table.items())
  len_encode_char=0
  for i in items_frequence:
    len_encode_char+=i[1]*len(huffman[i[0]])
  #Số ký tự phân biệt có trong văn bản:
  numofchar=len(items_frequence)
  #số bit cần có để biểu diễn đủ số ký tự này:
  numofbit=float(math.sqrt(numofchar))
  if numofbit-int(numofbit)==0:
    numofbit=int(numofbit)
  else:
    numofbit=int(numofbit)+1
  #tổng số bit để biểu diễn tất cả các kí tự trong file text văn bản:
  b=numofbit*len(text)
  return round(len_encode_char/b,2)
```
